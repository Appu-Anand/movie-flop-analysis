```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look better
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("tmdb_5000_movies.csv")
df.head()
```

```python
print("Rows and Columns:", df.shape)
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
columns_to_drop = ['homepage', 'keywords', 'production_companies', 'production_countries', 
                   'spoken_languages', 'status', 'tagline']

df.drop(columns=columns_to_drop, inplace=True)
df.head(3)
```

```python
df['profit'] = df['revenue'] - df['budget']

df = df.sort_values(by='profit')
df[['title', 'budget', 'revenue', 'profit']].head()
```

```python
flop_movies = df[df['profit'] < 0]

print("Total Flop Movies:", flop_movies.shape[0])
flop_movies[['title', 'budget', 'revenue', 'profit']].head()
```

```python
print("Duplicates:", df.duplicated().sum())

flop_movies.isnull().sum()
```

```python
flop_movies['genres'].value_counts().head(10)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=flop_movies, x='budget', y='revenue', alpha=0.6)
plt.title("Budget vs Revenue (Flop Movies)")
plt.xlabel("Budget")
plt.ylabel("Revenue")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(flop_movies['profit'], bins=30, kde=True, color='crimson')
plt.title("Distribution of Loss Amounts")
plt.xlabel("Profit (Negative = Loss)")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(data=flop_movies, x='runtime', y='profit')
plt.title("Runtime vs Profit (Flop Movies)")
plt.xlabel("Runtime")
plt.ylabel("Profit")
plt.show()

plt.figure(figsize=(10, 5))
sns.scatterplot(data=flop_movies, x='popularity', y='profit')
plt.title("Popularity vs Profit (Flop Movies)")
plt.xlabel("Popularity")
plt.ylabel("Profit")
plt.show()
```

```python
print("Average Rating of Flops:", flop_movies['vote_average'].mean())

plt.figure(figsize=(10, 5))
sns.histplot(flop_movies['vote_average'], bins=10, kde=True, color='orange')
plt.title("Vote Averages of Flop Movies")
plt.xlabel("IMDb Rating")
plt.show()
```

```python
## 🔍 Key Insights from the Analysis

- Over **300+ movies** in the dataset had **negative profits**, meaning they lost money at the box office.
- Most loss-making movies had **large budgets but poor returns**, suggesting overinvestment in low-performing titles.
- The **IMDb ratings** of flops were often between **5 and 6**, showing that bad reviews likely contributed to poor revenue.
- There was **no clear correlation** between runtime/popularity and financial success in flop movies.
- A few **high-budget disasters** stood out with losses exceeding $100M.

> 🎯 This project demonstrates my ability to clean data, perform exploratory analysis, and extract actionable insights using Python, Pandas, and Seaborn.
```
